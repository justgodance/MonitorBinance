```python
def find_nth(text, needle, n):
    start = text.find(needle)
    while start >= 0 and n > 1:
        start = text.find(needle, start+len(needle))
        n -= 1
    return start
```

```python
def greenTink():
    pricet = [0 for i in range(10)]
    volumet = [0 for i in range(10)]
    headers = {
        "Accept": "*/*",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8",
        "Cache-Control": "no-cache",
        "Connection": "keep-alive",
        "Content-Length": "123",
        "content-type": "application/json",
        "Host": "p2p.binance.com",
        "Origin": "https://p2p.binance.com",
        "Pragma": "no-cache",
        "TE": "Trailers",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"
        }

    data = {
      "asset": "USDT",
      "fiat": "RUB",
      "merchantCheck": False,
      "page": 1,
      "payTypes": ["Tinkoff"],
      "publisherType": None,
      "rows": 10,
      "tradeType": "BUY"
        }


    r = requests.post('https://p2p.binance.com/bapi/c2c/v2/friendly/c2c/adv/search', headers=headers, json=data)
    kur = r.text
    i = 1
    while i <10:
        
        try: 
            index0 = find_nth(kur, 'price', 3)
            pricet[0] = float(kur[index0+8:index0+13])
            index1 = find_nth(kur, 'price', 3+4*i)
            pricet[i] = float(kur[index1+8:index1+13])
        except ValueError:
            pass
        
        i+=1
    j=1
    while j<=10:
        index = find_nth(kur, 'minSingleTransAmount', j)
        minvalue = kur[index+23:index+33] 
        if minvalue.find('"') == -1 :
            try: 
                volumet[j-1] = float(minvalue)
            except ValueError:
                pass
        else:
            try:
                val=minvalue.split('"',1)[0]
                volumet[j-1] =float(val) 
            except ValueError:
                pass
        j+=1
#print(float(kur[index1+8:index1+13]))
    price =list(pricet)
    volume = list(volumet)
    return (price, volume)
```

```python
def gar(driver):    
    driver.refresh()
    soup = BeautifulSoup(driver.page_source, "xml")
    table = soup.find_all('td',{'class':"price col-xs-6 overflow-aut"})#табл с куурсом
    #выделение курсов и объемов usdt
    i=100
    while i < 105:
        priceg[i-100]=table[i].text
        i+=1
    resultprice = [float(item) for item in priceg]
    return (resultprice)
```

```python
def best():
    page = requests.get("https://www.bestchange.ru/tether-trc20-to-tinkoff.html")
    soup = BeautifulSoup(page.text, 'html.parser')
    course = soup.find_all('td',{'class':"bi"})[:6] #course values
    names = soup.find_all('a',{'rel':"nofollow"})[:3] #link values
    check = soup.find_all('span',{'onclick':"stopBubbling(event)"})[:6] #check for manual
    checked = []
    p=0
    for item in check:
        if p%2!=0:
            checked.append(item.text)
        p+=1
    values = [] #final array
    for item in course: #getting course values
        string = item.text.split()
        if float(string[0])!=1:
            values.append(string[0])
    for item in names: #getting link values
        values.append(item['href'])
    for i in range(3):
        if checked[i].find("работает в ручномили")==-1:
            return values
        else:
            return 0
```
